#### Simulating 10,000 hours

```python
import numpy as np
import random
    
# X = State: 0: Growing, 1: Sequencing, 2: Dividing
X = [0]
N = 10000

def bernoulli_rv(p):
    """  
    Do a Bernoulli trial with rate of success of p (0,1)
    """
    u = random.uniform(0,1)
    if u<p:
        return 1
    else:
        return 0


for i in range(N):
    if X[-1] == 0:
        change_to_S = bernoulli_rv(1/10)
        if change_to_S:
            X.append(1)
        else:
            X.append(0)
    elif X[-1] == 1:
        change_to_D = bernoulli_rv(1/8)
        if change_to_D:
            X.append(2)
        else:
            X.append(1)
    else:
        change_to_G = bernoulli_rv(2/5)
        if change_to_G:
            X.append(0)
        else:
            X.append(2)

print(f'Time in G = {X.count(0)}')
```
