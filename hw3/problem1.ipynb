#### Problem 1
##### Sample from $f$ using optimal value of $c(a)$

```python
import numpy as np
import plotly.graph_objects as go

# Functions
c = lambda a: np.exp(-a)/(3*a**2*(1-a))
f = lambda x: 1/3*x*(1+x)*np.exp(-x)
def g(x,a,c):
    return c*a**2*x*np.exp(-a*x)

# Calculating values
a_star = (6-np.sqrt(108))/(-6)
C = c(a_star)
x = np.linspace(0.5,50,200)
fx = f(x)
gcx = g(x,a_star,C)

# Creating plots
fig = go.Figure()
f_trace = go.Scatter(x=x,y=fx,mode='lines',name='f(x)')
g_trace = go.Scatter(x=x,y=gcx,mode='lines',name='c(a)g_a(x)')
fig.add_trace(f_trace)
fig.add_trace(g_trace)
fig.update_layout(title='f(x) and c(a)g(x)',xaxis_title='x',yaxis_title='f(x),cg(x)')
fig.show()
```
